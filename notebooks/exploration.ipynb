```python
#Feature Engineering 
#Feature 4.1 — Threat Severity Score

df["severe attack"]= df["Financial Loss (in Million $)"] + df["Number of Affected Users"]
```

```python
df["severe attack"].head()
```

```python
#Feature 4.2 — Loss Per User
df["loss per user"]= df["Financial Loss (in Million $)"] / df["Number of Affected Users"]
```

```python
df["loss per user"].head()
```

```python
#Feature 4.3 — Resolution Efficiency
df["Resolution Efficeincy"]= df["Incident Resolution Time (in Hours)"] / df["Number of Affected Users"]
```

```python
df["Resolution Efficeincy"].head()
```

```python
#Feature 4.4 — Risk Level
def risk_level(loss):

    if loss > 66:
        return "High Risk"

    elif loss > 33:
        return "Medium Risk"

    else:
        return "Low Risk"
    
    
```

```python
df["Risk Level"] = df["Financial Loss (in Million $)"].apply(risk_level)
```

```python
df[["Financial Loss (in Million $)", "Risk Level"]].head()
```

# Feature 5.4 — ML Insights

1. Number of Affected Users and Incident Resolution Time show an extremely weak positive correlation (0.005893).

2. Financial Loss and Incident Resolution Time show an extremely weak negative correlation (-0.012671).

3. Financial Loss and Number of Affected Users show almost no correlation (0.001787).

4. No strong linear relationship exists between the numerical features.

5. The numerical features are largely independent and are not good predictors of one another based on correlation analysis.

```python
#Phase 6 — Data Transformation
#feature 6.1 renaming the feature coulumn 
df.rename(columns={"Financial Loss (in Million $)":"financial_loss"})
```
