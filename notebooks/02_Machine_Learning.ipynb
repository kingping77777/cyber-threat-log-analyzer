project preprocessing phase 4 

```python
X.select_dtypes(include="object").columns
```

```python
# Create X
X = df[[
    "Country",
    "Year",
    "Attack Type",
    "Target Industry",
    "Number of Affected Users",
    "Attack Source",
    "Security Vulnerability Type",
    "Defense Mechanism Used",
    "Incident Resolution Time (in Hours)"
]]

Y = df["Financial Loss (in Million $)"]
```

```python
#coverting all the coloumns to the one hot coding 
X = pd.get_dummies(
    X,
    columns=[
        "Country",
        "Attack Type",
        "Target Industry",
        "Attack Source",
        "Security Vulnerability Type",
        "Defense Mechanism Used"
    ]
)
```

```python
# THEN split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
X.info()
```

feature scaling phase 5 

```python
#phase 5 #feature 5.1
#feature scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

```

```python
print(X_train_scaled[:5])
```

```python
#feature5.2 

from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

X_train_minmax = minmax.fit_transform(X_train)

X_test_minmax = minmax.transform(X_test)
```

```python
print(X_train_minmax[:5])
```

```python
#phase 6 liner regresion 
#feature 6.8
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled,Y_train)
#THE MODEL HAS BEEN TRAINED 
```

```python
prediction = model.predict(X_test_scaled)
```

```python
print(prediction[:5])

```

```python
print(Y_test.iloc[:5])
```

```python
#model evaluation r^2 #feature 6.12 
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, prediction)
print(r2)
```

```python
#feature6.13
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(Y_test, prediction)
mean_absolute_error(Y_test,prediction)
print(mae)
```

```python
#feature 6.14 Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, prediction)
print(mse)
```

```python
#feature 6.15 Root Mean Squared Error (RMSE)
import numpy as np
rmse = np.sqrt(mse)
print(rmse)
```
